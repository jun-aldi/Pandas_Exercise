```python
# Membagi Data Frame menjadi dua secara acak
```

```python
import pandas as pd, numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
#cek shape data frame

df.shape
```

```python
#proprsi :
# 70% df 1 30% df 2
proporsi = 0.7

#df.sample merupakan method mengambil baris secara acak
#ditampung di df 1
df_1 = df.sample(frac=proporsi)

#df hapus berdasrkan datframe yang sudah diadopsi pada df 1
df_2 = df.drop(df_1.index)

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
```
