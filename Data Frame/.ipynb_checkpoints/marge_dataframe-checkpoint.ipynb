```python
#Menggabungkan dua Pandas Data Frame
```

```python
import pandas as pd, numpy as np
```

```python
n_rows = 5
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns = cols)

df
```

```python
df1 = df.copy(deep=True)
#baris 1 dan 4 kita buang
df1 = df1.drop([1, 4])
df1
```

```python
df2 = df.copy(deep=True)
#baris 1 kita buang
df2 = df2.drop([0, 3])
df2
```

```python
#mengabungkan dua data frame
```

```python
df_inner = pd.merge(df1, df2, how='inner')
#how inner = BARIS MENGGABUNGKAN DUA BUAH DATA FRAME YANG TERDAPAT DATA FRAME KEDUANNYA
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer')
#how outer = BARIS MENGGABUNGKAN DUA BUAH DATA FRAME YANG DUPLIKASI HANYA SATU SAJA
df_outer
```
