```python
#Seleksi kolom dan baris pada Data Frame menggunakan loc
```

```python
import pandas as pd, numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns = cols)

df
```

```python
#Seleksi kolom dan baris dengan loc
```

```python
#loc

df.loc[[0,3,4],['B','E']]

#bagian pertama mengatur baris
#bagian kedua untuk kolom
#dipisahkan tanda koma
#contoh tsb ingin mengakses index ke 0, 3, 4, lalu kita bungkus dalam suatu list []
#lalu pada kolom B dan E
```

```python
#Seleksi baris dengan kondisi
```

```python
df.loc[df['B']>10, ['B', 'D', 'E']]

#nilai kolom B nya diatas 10
#hanya menampilkan kolam B D E saja
```

```python
#slicing data frame dengan loc
```

```python
df.loc[0:4, 'B':'D']
```
