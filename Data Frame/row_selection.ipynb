```python
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,5, size=(n_rows, n_cols))
                  ,columns=cols)
df
```
