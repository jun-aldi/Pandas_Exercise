```python
import pandas as pd
import numpy as np
```

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10, size=(n_rows,n_cols)),
                 columns=cols)
df
```

```python
#membalik urutan kolom
#df.loc["mengatur row", "mengatur kolom"]
#: tanda untuk menyertkan keseluruhanbaris
df.loc[:, ::-1]
```

```python
#mambalik urutan baris
#baris urutan pertama tidak perlu menyertakan kolom
df.loc[::-1]
```

```python
#mambalik urutan baris dengan re-indexin
#baris urutan pertama tidak perlu menyertakan kolom
#drop=true index yg lama akan dibuang atau re-indexing
df.loc[::-1].reset_index(drop=True)
```
