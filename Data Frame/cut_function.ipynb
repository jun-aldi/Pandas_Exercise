```python
#Konversi nilai numerik ke dalam kategori pada Data Frame
```

```python
import pandas as pd, numpy as np
```

```python
n_rows = 10
n_cols = 1
cols=('usia',)

df = pd.DataFrame(np.random.randint(1,99, size=(n_rows, n_cols)),
                 columns=cols)

df
```

```python
#mengelompokan nilai kelompk ke beberapa kategori
#dengan function .cut
```

```python
#0 - 18 anak
#18 - 65 dewasa
#65 - 99 manula
df['kelompk usia']=pd.cut(df['usia'],
                         bins = [0, 18, 65, 99],
                         labels =['anak', 'dewasa', 'manula'])
df
```
